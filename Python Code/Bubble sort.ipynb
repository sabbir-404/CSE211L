```python
def bubble_sort(number):
    number_list = [int(digit) for digit in str(number)]
    number_length = len(number_list)
    count = 0
    for i in range(number_length - 1, 0, -1):
        for j in range(i):
            if number_list[j] > number_list[j + 1]:
                number_list[j], number_list[j + 1] = number_list[j + 1], number_list[j]
                count += 1
                print(count, end = '. ')
                print(' '.join([str(x) for x in number_list]))
                print()
                
    return (''.join([str(x) for x in number_list]))

input_number = 5638
sorted_number = bubble_sort(input_number)
print(sorted_number)
```

```python
import string

def word_printer(text):
    tmp = 'abcdefghijklmnopqrstuvwxyz'
    word_lst = [str(txt) for txt in text.lower()]
    alphabets = [str(txt) for txt in tmp]
    
    for char in word_lst:
        for alph in alphabets:
            if char != alph:
                print(alph)
            else:
                print(char)
    
    print()

word = 'asdfg'

for i in range(len(word)):
    if word[i].isalpha():
        print(word_printer(word[i]), end = '')
```
